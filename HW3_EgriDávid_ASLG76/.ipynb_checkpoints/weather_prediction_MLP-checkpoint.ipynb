# Import the packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook

from sklearn.preprocessing import StandardScaler, MinMaxScaler

from keras.models import Sequential
from keras.layers.core import Activation, Dense
from keras.optimizers import SGD
from keras.callbacks import ModelCheckpoint, EarlyStopping, Callback
from keras.models import load_model


from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

import warnings
warnings.filterwarnings("ignore")
```

# Load the data from the .csv file

```python
data = pd.read_csv('data.csv', index_col=0, delimiter = '\t', header=0)
data.index = pd.to_datetime(data.index)
```

Plot the loaded data

```python
fig, ax = plt.subplots()
data.plot(x_compat=True, ax=ax, figsize=(9,5), grid=5, marker='.')
```

# Data preparation for the MPL

Collect the previous days and drop those that do not have enough previous days

```python
# get the shifted days
for i in range(1,5):
    data['Tempearature_shift_' + str(i)] = data.Temperature.shift(i)

# Drop Nan-s
data = data.dropna()
# Show the first rows
data.head()
```

Do the train, valid, test split

```python
valid_split = 0.2
test_split = 0.1
nb_samples = data.shape[0]

data_train = data.iloc[:int(nb_samples*(1-valid_split-test_split))] 
data_valid = data.iloc[int(nb_samples*(1-valid_split-test_split)):int(nb_samples*(1-test_split))]
data_test  = data.iloc[int(nb_samples*(1-test_split)):]
```

Use the standard scaler
* I used this instead of the MinMax beacause the negative tempareture values
* Because of the periodicity in temperature I fitted the scaler for every part

```python
inputs = data.columns[1:]
output = data.columns[0]

data_train[inputs] = StandardScaler().fit_transform(data_train[inputs])
data_valid[inputs] = StandardScaler().fit_transform(data_valid[inputs])
data_test[inputs] = StandardScaler().fit_transform(data_test[inputs])

data_train[output] = StandardScaler().fit_transform(data_train[output].values.reshape(-1,1))
data_valid[output] = StandardScaler().fit_transform(data_valid[output].values.reshape(-1,1))
data_test[output] = StandardScaler().fit_transform(data_test[output].values.reshape(-1,1))
```

Use early stopping and check point

```python
es = EarlyStopping(patience=30, verbose=1)
mcp = ModelCheckpoint(filepath='weights.hdf5', verbose=1, save_best_only=True)
```

Own history to store the losses during the training

```python
class TrainingHistory(Callback):
  def on_train_begin(self, logs={}):
    self.losses=[]
    self.valid_losses =[]
    self.accs = []
    self.valid_accs = []
    self.epoch=0
    
  def on_epoch_end(self, epoch,logs={}):
    self.losses.append(logs.get('loss'))
    self.valid_losses.append(logs.get('val_loss'))
    self.accs.append(logs.get('acc'))
    self.valid_accs.append(logs.get('val_acc'))
    self.epoch += 1
    
history = TrainingHistory()
```

Build up the modell

```python
model = Sequential()
model.add(Dense(20, input_shape=(4,), use_bias=True, activation='relu'))
model.add(Dense(20, use_bias=True, activation='relu'))
model.add(Dense(5, use_bias=True, activation='relu'))
model.add(Dense(1, use_bias=False, activation='linear'))

model.compile(loss='mse', optimizer='adam', metrics=['mae'])

model.summary()
```

Train the modell

```python
hst = model.fit(data_train[inputs].values, data_train[output].values,
                   epochs=10000000,
                   batch_size=8,
                   verbose=2,
                   validation_data=(data_valid[inputs].values, data_valid[output].values),
                   callbacks=[mcp,es,history])
```

Plot the error during the training

```python
plt.figure(figsize=(9, 5))

plt.plot(np.arange(history.epoch), history.losses, color='g',  
         label="Error on the training data")
plt.plot(np.arange(history.epoch), history.valid_losses, color='r',
         label="Error on the validation data")

plt.title("Error during training")
plt.legend(loc='upper right')
plt.xlabel("Number of epochs")
plt.ylabel("Error")

plt.grid(True)
plt.show()
```

Load back the best modell and predict the test data

```python
model = load_model('weights.hdf5')
data_test['predictions'] = model.predict(data_test[inputs].values)
data_test.plot(y=['predictions', 'Temperature'], x_compat=True, figsize=(9,5), grid=5)
```
