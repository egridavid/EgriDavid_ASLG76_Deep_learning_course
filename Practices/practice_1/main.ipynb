# Milyen kategorikus változók lehetnek: color, director_name, actor_1_name, actor_2_name, actor_3_name, language, country, content_rating

```python
categorical_features = ['color', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'language', 'country', 'content_rating']
n_items = len(df)
```

# Amelyik kategórinál az összes elem kevesebb mint 2% tartozik ott OTHER kategóriába raktuk őket

```python
for feat in categorical_features:
    
    print("A %s feldolgozása. Egyedi mezők száma: %d." % (feat, df[feat].nunique()))
    
    if df[feat].nunique() < 50:

        print(df[feat].unique())
        print(df.groupby(feat).count().iloc[:,0])
        
        # Itt rakjuk át a OTHER kategóriába a cuccokat
        for feat_value in df[feat]:
            if (len(df[df[feat] == feat_value]) / n_items < 0.02):
                print("A %s kategória hozzáadása az OTHER kategoriához" % feat_value)
                df[feat] = df[feat].apply(lambda x: "OTHER" if x == feat_value else x, 1)
```

```python
df = pd.get_dummies(df, columns=['color'])
```
