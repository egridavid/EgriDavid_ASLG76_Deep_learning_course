## Get the image database for the 5. and 6. task

* CIFAR-10 dataset is used for the exercises

* Cite: [Learning Multiple Layers of Features from Tiny Images], Alex Krizhevsky, 2009.

[Learning Multiple Layers of Features from Tiny Images]: https://www.cs.toronto.edu/~kriz/learning-features-2009-TR.pdf

```python
# data batch file from the used dataset
file = Path(r'/Users/egridavid.bp/Desktop/Deep learning a gyakorlatban/homework_1/data/cifar-10-batches-py/data_batch_1')

# a function that unpickle the dateset object
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

# the data set as a dictionary
source = unpickle(file)

# fixed random seed
np.random.seed(666)
# get five random int to get five random images from the dataset
indexes = np.array(np.random.rand(5)*source[b'data'].shape[0], dtype=np.uint8)

database = {}
for i in indexes:
    database[source[b'filenames'][i].decode("utf-8")[:-13]] = source[b'data'][i].reshape((3,32,32)).transpose([1,2,0])
```

# 5: read 5 32x32 pixel RGB images, print out the RGB chanell data, plot the images

Plot the database

```python
fig = plt.figure(figsize=(9,9))
axes = fig.subplots(nrows=4, ncols=5)

for key in database.keys():
    image = database[key]
    i = list(database.keys()).index(key)
    axes[0,i].imshow(image) # original image
    axes[0,i].set(title = key) # get the image content from the databse
    axes[1,i].imshow(image[:,:,0], cmap = 'Reds') # red RGB channel
    axes[2,i].imshow(image[:,:,1], cmap = 'Greens') # green RGB channel
    axes[3,i].imshow(image[:,:,2], cmap = 'Blues') # blue RGB channel

plt.tight_layout()
```

# 6: Calculate the mean and the deviation of the pixels for the five image in every channel. Convert these to a 0 mean, 1 deviation dataset

Define a myscaler function

```python
def myscaler(image):
    temp = np.empty((32, 32, 3))
    for i in range(3):
        normalized = (image[:,:,i] - image[:,:,i].mean()) / image[:,:,i].std()
        temp[:,:,i] = normalized
    return temp
```

Get the normalized database

```python
normalized_database = {}
for key in database.keys():
    normalized_database[key] = myscaler(database[key])
```

Example for the normalized database

```python
key = list(database.keys())[3] # choose a random key to get an example image

image = database[key]
normalized_image = normalized_database[key]

for i in range(3):
    print('chanel:', i, '\n\n')
    print('RGB values:')
    print(image[:,:,i],'\n')
    print('mean:', image[:,:,i].mean(), 'std:', image[:,:,i].std(),'\n')
    print('normalized values:')
    print(normalized_image[:,:,i],'\n\n')
```

## 7: Read a soundfile and plot the spektrogram

```python
url = 'https://freewavesamples.com/files/Claves.wav' # the url of the source file

r = requests.get(url, allow_redirects=True) # download the file from the url to a file
open('data.wav', 'wb').write(r.content)

fs, data = wavfile.read('data.wav') # read the wav file
plt.specgram(data[:,0], Fs=fs) # plot the specgram, stereo wav file --> data[:,0] indexing is needed
```
